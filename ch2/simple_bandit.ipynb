```python
import numpy as np
import matplotlib.pyplot as plt
from utils import e_greedy_bandit
```

```python
k = 10
max_steps = 1000
nb_runs = 2000
avg_reward = []
eps = [0, 0.01, 0.1]
colors = ['Blue', 'Red', 'Green']
labels = ['eps=0 (greedy)', 'eps=0.01', 'eps=0.1']
```

```python
for e in eps:
    avg_reward.append(np.zeros(max_steps))
    for j in range(1, nb_runs+1):
        avg_reward[-1] += 1/j*(e_greedy_bandit(k, max_steps, e) - avg_reward[-1])
```

```python
fig, ax = plt.subplots(figsize=(16, 8))
for r,c,l in zip(avg_reward, colors, labels):
    ax.plot(r, color = c, label = l)
plt.ylabel('Average reward')
plt.xlabel('Steps')
plt.legend(loc="lower right")
plt.ylim([0, 1.5])
plt.savefig('avg_bandit.pdf')
plt.show()
```
